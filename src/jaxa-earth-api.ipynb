```python
from jaxa.earth import je
```

```python
data = je.ImageCollection(ssl_verify=True)\
    .filter_date()\
    .filter_resolution()\
    .filter_bounds()\
    .select()\
    .get_images()

img = je.ImageProcess(data)\
    .show_images()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img.raster.img[0])
plt.show()
```

Filter by date and bbox

```python
kw   = ["Aqua", "LST", "half-monthly"]
dlim = ["2021-01-01T00:00:00","2021-01-01T00:00:00"]
ppu  = 5
bbox = [110, 20, 160, 50]
```

```python
collections, bands = je.ImageCollectionList(ssl_verify=True)\
                       .filter_name(keywords=kw)
```

```python
data = je.ImageCollection(collection=collections[0], ssl_verify=True)\
         .filter_date(dlim=dlim)\
         .filter_resolution(ppu=ppu)\
         .filter_bounds(bbox=bbox)\
         .select(band=bands[0][0])\
         .get_images()

img = je.ImageProcess(data)\
        .show_images()
```

# GCOM-C LST

```python
collection = "JAXA.G-Portal_GCOM-C.SGLI_standard.L3-LST.daytime.v3_global_half-monthly"
dlim       = ["2022-11-01T00:00:00","2023-10-31T00:00:00"]
band       = "LST"
```

```python
geoj_path = "vietnam_country.geojson"
geoj = je.FeatureCollection().read(geoj_path).select()
```

```python
data = je.ImageCollection(collection, ssl_verify=True)\
         .filter_date(dlim)\
         .filter_resolution()\
         .filter_bounds(geoj=geoj[0])\
         .select(band)\
         .get_images()
```

```python
img = je.ImageProcess(data)\
        .show_images()
```

```python
# Process and show an image
img = je.ImageProcess(data)\
        .calc_spatial_stats()\
        .show_spatial_stats()
```
